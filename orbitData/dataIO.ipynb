```python
import pandas as pd
import numpy as np
import time                     #python 自带
from astropy.time import Time   #astropy 
```

```python
path = './LISA-Like-Orbit/20280322_LISA_2p5Mkm'  #LISA armlength 2.5 Mkm

sunP = pd.read_csv(path + '/SunP.dat', header=None, sep=' ', skipinitialspace=True)
sunP.columns = ['sunPX','sunPY','sunPZ']

earthP = pd.read_csv(path + '/Earth.dat', header=None, sep=' ', skipinitialspace=True)
earthP.columns = ['earthPX','earthPY','earthPZ']

scp1 =  pd.read_csv(path + '/SCP1.dat', header=None, sep=' ', skipinitialspace=True)
scv1 =  pd.read_csv(path +'/SCV1.dat', header=None, sep=' ', skipinitialspace=True)
scp1.columns = ['scp1X','scp1Y','scp1Z']
scv1.columns = ['scv1X','scv1Y','scv1Z']

scp2 =  pd.read_csv(path + '/SCP2.dat', header=None, sep=' ', skipinitialspace=True)
scv2 =  pd.read_csv(path + '/SCV2.dat', header=None, sep=' ', skipinitialspace=True)
scp2.columns = ['scp2X','scp2Y','scp2Z']
scv2.columns = ['scv2X','scv2Y','scv2Z']

scp3 =  pd.read_csv(path + '/SCP3.dat', header=None, sep=' ', skipinitialspace=True)
scv3 =  pd.read_csv(path + '/SCV3.dat', header=None, sep=' ', skipinitialspace=True)
scp3.columns = ['scp3X','scp3Y','scp3Z']
scv3.columns = ['scv3X','scv3Y','scv3Z']

timeVec = np.linspace(2461853.0, 2461853.0 + len(sunP) -1, len(sunP)) #begin, end, n
timeVecJD = Time(timeVec, format='jd', scale='tcb')  #tcb : Barycentric Coordinate Time (TCB)
timeJD = pd.DataFrame(timeVecJD, columns = ['timeJD'])
#建立一个astropy time 对象
#Time.SCALES
#timeVecJD.tdb

orbitData = pd.concat([timeJD, sunP, earthP, scp1, scv1, scp2, scv2, scp3, scv3], axis = 1 )  
#建立一个dataframe对象

#timeVecJD = Time(timeVec, format='jd', scale='tcb')  #tcb : Barycentric Coordinate Time (TCB)


orbitData.to_csv('orbitDataLISA.dat')# columns=['JDTimeTCB']  #导出csv文件

data  =  pd.read_csv('orbitDataLISA.dat',index_col = 0)  #读取csv文件
data
```
