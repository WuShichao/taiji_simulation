```python
import pandas as pd
import numpy as np
import time                     #python 自带
from astropy.time import Time   #astropy 
```

```python
timeVec = pd.read_csv('./TimeSeries.dat', header=None)
timeVec.columns = ['JDTimeTCB']

earthP = pd.read_csv('./Earth.dat', header=None, sep=' ', skipinitialspace=True)
earthP.columns = ['earthPX','earthPY','earthPZ']

sunP = pd.read_csv('./SunP.dat', header=None, sep=' ', skipinitialspace=True)
sunP.columns = ['sunPX','sunPY','sunPZ']

scp1 =  pd.read_csv('./SCP1.dat', header=None, sep=' ', skipinitialspace=True)
scv1 =  pd.read_csv('./SCV1.dat', header=None, sep=' ', skipinitialspace=True)
scp1.columns = ['scp1X','scp1Y','scp1Z']
scv1.columns = ['scv1X','scv1Y','scv1Z']

scp2 =  pd.read_csv('./SCP2.dat', header=None, sep=' ', skipinitialspace=True)
scv2 =  pd.read_csv('./SCV2.dat', header=None, sep=' ', skipinitialspace=True)
scp2.columns = ['scp2X','scp2Y','scp2Z']
scv2.columns = ['scv2X','scv2Y','scv2Z']

scp3 =  pd.read_csv('./SCP3.dat', header=None, sep=' ', skipinitialspace=True)
scv3 =  pd.read_csv('./SCV3.dat', header=None, sep=' ', skipinitialspace=True)
scp3.columns = ['scp3X','scp3Y','scp3Z']
scv3.columns = ['scv3X','scv3Y','scv3Z']

orbitData = pd.concat([timeVec, sunP, earthP, scp1, scv1, scp2, scv2, scp3, scv3], axis = 1 )  
#建立一个dataframe对象

timeVecJD = Time(timeVec, format='jd', scale='tcb')  #tcb : Barycentric Coordinate Time (TCB)
#建立一个astropy time 对象
#Time.SCALES
#timeVecJD.tdb

orbitData.to_csv('orbitData.dat')# columns=['JDTimeTCB']  #导出csv文件
```
